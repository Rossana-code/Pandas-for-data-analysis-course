#### Remplir les NA

* Traiter les NAN si leur nombre est superieur à 30

```python
for colonne in df.columns:
    if df[colonne].isnull().sum() > 30:
        # Appliquer le traitement de votre choix ici. 
        # Par exemple, remplacer les NaN par la moyenne de la colonne :
        df[colonne] = df[colonne].fillna(df[colonne].mean())
        # Ou pour supprimer les lignes contenant des NaN dans cette colonne :
        # df = df[df[colonne].notna()]
```

```python
df.isna().sum()
```

* Traiter toutes les colonnes contenant les NAN

```python
for i in range(0,df.shape[1]-1):
    if np.dtype(df.iloc[:,i]) == 'float64':
        df.iloc[:,i].fillna(np.mean(df.iloc[:,i]), inplace=True)
    else :
        df.iloc[:,i].fillna(df.iloc[:,i].value_counts().idxmax(), inplace=True)
```
