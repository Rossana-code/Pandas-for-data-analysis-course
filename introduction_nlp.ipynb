Chacune de ces lignes de code utilise la fonction `download` de la bibliothèque NLTK (Natural Language Toolkit) pour télécharger des ressources spécifiques utilisées dans le traitement du langage naturel. Voici ce que chacune fait :

1. `nltk.download('averaged_perceptron_tagger')`: 
   - Cette ligne télécharge le "Averaged Perceptron Tagger" de NLTK. 
   - Un "tagger" en traitement du langage naturel est un outil qui attribue des étiquettes aux mots d'une phrase, généralement pour identifier leur rôle grammatical (comme nom, verbe, adjectif, etc.). 
   - L'"Averaged Perceptron Tagger" est un type de tagger basé sur l'algorithme de perceptron, un type de modèle d'apprentissage automatique pour la classification.

2. `nltk.download('maxent_ne_chunker')`: 
   - Cette commande télécharge le "Maximum Entropy Named Entity Chunker" de NLTK. 
   - Le "Named Entity Chunker" est utilisé pour identifier des groupes de mots dans un texte qui représentent des entités nommées (comme les noms de personnes, d'organisations, ou de lieux).
   - Le terme "Maximum Entropy" se réfère à l'algorithme utilisé pour la classification, qui est basé sur le principe de maximisation de l'entropie, un concept de la théorie de l'information et de la statistique.

3. `nltk.download('words')`: 
   - Cette ligne télécharge une liste de mots anglais de NLTK.
   - Cette liste est souvent utilisée comme une sorte de dictionnaire de base pour diverses tâches en NLP, comme la vérification orthographique, ou comme une liste de référence pour identifier les mots dans le traitement de texte.
   
En résumé, ces téléchargements fournissent des outils essentiels pour la tokenisation, le taggage de parties du discours, la reconnaissance d'entités nommées, et d'autres tâches de base en NLP.

```python
from nltk import word_tokenize , pos_tag , ne_chunk
# pos_tag : permet de reconnaitre les lieux et ne_chunk des personnes
```

1. **`word_tokenize`** :
   - `word_tokenize` est une fonction de NLTK utilisée pour diviser une chaîne de caractères (texte) en une liste de mots individuels. 
   - C'est une étape fondamentale dans de nombreuses tâches de traitement du langage naturel (NLP), car elle convertit un bloc de texte en tokens (généralement des mots et des signes de ponctuation) qui peuvent être analysés individuellement.

2. **`pos_tag`** :
   - `pos_tag` est utilisée pour effectuer le taggage de parties du discours. 
   - Elle prend une liste de mots (tokens) et attribue à chacun une étiquette de partie du discours, comme nom (NN), verbe (VB), adjectif (JJ), etc. 
   - Cette fonction est utile pour comprendre la structure grammaticale d'une phrase et pour des tâches comme l'analyse de sentiments, la reconnaissance d'entités nommées, et la traduction automatique.

3. **`ne_chunk`** :
   - `ne_chunk` est utilisée pour la reconnaissance d'entités nommées (Named Entity Recognition, NER). 
   - Elle identifie et classe certaines parties d'un texte comme étant des entités nommées, telles que les noms de personnes, les organisations, les lieux, etc. 
   - Par exemple, dans la phrase "George Washington était le premier président des États-Unis", `ne_chunk` pourrait identifier "George Washington" comme une personne (PERSON) et "États-Unis" comme une entité géopolitique (GPE).

En somme, cette ligne de code importe trois outils puissants de NLTK pour le traitement de texte : la tokenisation de mots, le taggage de parties du discours, et la reconnaissance d'entités nommées. Ces outils sont essentiels pour diverses applications en NLP.

```python
token = word_tokenize(u'John works at Linkedin in Paris in France')
print(ne_chunk(pos_tag(token)))

# GPE : lieu
```

### Bag of words

La différence principale entre le modèle "Bag of Words" (BoW) et la Binarisation dans le traitement du langage naturel (NLP) réside dans la manière dont les mots sont représentés et utilisés pour analyser les textes. Voici une explication détaillée de chacun :

1. **Bag of Words (BoW) :**
   - Le modèle BoW est une représentation simplifiée du texte qui transforme le texte en un ensemble de nombres. 
   - Dans ce modèle, tous les mots uniques dans l'ensemble du texte (ou corpus) sont considérés, et chaque mot est traité comme une caractéristique distincte.
   - La caractéristique principale de BoW est qu'il compte la fréquence d'occurrence de chaque mot dans un document. Cela signifie que pour chaque document, le modèle BoW enregistre combien de fois chaque mot apparaît.
   - Par exemple, si le mot "chat" apparaît 3 fois dans un document, l'entrée correspondante dans la représentation BoW de ce document serait quelque chose comme "chat: 3".
   - Cependant, cette méthode ne tient pas compte de l'ordre des mots ni du contexte dans lequel ils sont utilisés.

2. **Binarisation :**
   - La binarisation est une variante simplifiée du modèle BoW. Au lieu de compter la fréquence, elle se contente de noter la présence ou l'absence d'un mot dans un document.
   - Dans cette représentation, chaque mot est à nouveau traité comme une caractéristique distincte. Cependant, au lieu de compter combien de fois un mot apparaît, la binarisation indique simplement si le mot est présent (1) ou absent (0) dans le document.
   - Par exemple, si le mot "chat" apparaît dans un document (peu importe combien de fois), l'entrée correspondante serait "chat: 1". Si le mot "chat" n'apparaît pas, l'entrée serait "chat: 0".
   - Comme pour le modèle BoW, la binarisation ne prend pas en compte l'ordre des mots ou leur contexte.

En résumé, le modèle BoW se concentre sur la fréquence des mots dans les documents, tandis que la binarisation se concentre uniquement sur leur présence ou leur absence. La binarisation peut être vue comme une version simplifiée de BoW, où l'information sur la fréquence est réduite à une valeur binaire.

* Créer un bag of words (sac de mots)

```python
from sklearn.feature_extraction.text import CountVectorizer

# Création d'un sac vide
vectorizer = CountVectorizer()
print(vectorizer)
print('--------------')

# Création d'un corpus 
corpus = ['un jour ici et là', 'demain mais pas ici', 'la la la lere']
print(corpus)
print('--------------')

# Calcul du sac
vectorizer.fit(corpus)
bow = vectorizer.transform(corpus)
print(bow)
print('--------------')

# Affichage des mots
feature_names = vectorizer.get_feature_names_out()  # Utilisation de get_feature_names_out()
print(feature_names)
```

```python
# ici on a trois documents : 
# 'un jour ici et là', 'demain mais pas ici', 'la la la lere'
# (2, 4)	3  dans document 2, le 4em mot (les espace sont comptés), 
# il apparait 3 fois

# (2, 5)	1 : document 2, 5em mot, 1 fois
```

### TF-IDF

* Exemple

```python
from sklearn.feature_extraction.text import CountVectorizer # compter
from sklearn.feature_extraction.text import TfidfTransformer # poids

corpus = ['un jour ici et là ' , 'demain mais pas la la la ici' , 'la la la la la la lere']

vectorizer = CountVectorizer()

# comptage
counts = vectorizer.fit_transform(corpus)
counts
```

```python
# calcul TFIDF à partir du comptage
transformer = TfidfTransformer()
transformer.fit_transform(counts)
# IFIDF diminue la ponderatin des presents dans plusieurs documents.
# Exemple les mots "ici" & "la"

transformer.idf_  # les poids 
# len(transformer.idf_)
```

```python
# Création du dictionnaire à partir des noms des caractéristiques et des scores IDF
dic = dict(zip(vectorizer.get_feature_names_out(), transformer.idf_))

# Création du DataFrame à partir du dictionnaire
df = pd.DataFrame.from_dict(dic, orient="index", columns=['IDF'])
print(df)
```

### Diagramme en barres des mots les plus fréquents

```python
from nltk.probability import FreqDist
```

```python
# Tokenisation du texte
# Supposons que 'all_text' contient votre texte

tokens = nltk.word_tokenize(all_text)

# Calcul de la fréquence des mots
fdist = FreqDist(tokens)

# Préparation des données pour le graphique
words = [word for word, freq in fdist.most_common(20)]
print(words)
print("------------------")

frequency = [freq for word, freq in fdist.most_common(20)]
print(frequency)
print("------------------")

# Création du graphique en barres
plt.figure(figsize=(6, 4))
plt.bar(words, frequency)
plt.xlabel('Mots')
plt.ylabel('Fréquence')
plt.xticks(rotation = 45)
plt.title('Top 20 des mots les plus fréquents')
plt.show()
```
