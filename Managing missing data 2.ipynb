* dropna

```python
df2 = df.dropna()  # drop every row having NAN
df2.shape
```

* all

```python
df3 = df.dropna(how = 'all',subset = ['T3'])   
print(df3.shape)
df3.head()
# all : if all value of a row are missing, drop it
```

* any

```python
df4 = df.dropna(how = 'any', subset = ['T3'])   
print(df4.shape)
df4.head()
# any : if just one value of a row is missing, drop the row
```
